```python
import math

fig, ax = plt.subplots(figsize=(20, 16))

only_kernel = [0.9343, 0.9348, 0.912, 0.6352, math.nan]
only_shift = [0.9343, 0.9309, 0.9214, 0.8145, math.nan]
shfit_and_fft = [0.9343, 0.9251, 0.8901, 0.9008, 0.7971]
combined = [0.9343, 0.9348, 0.9267, 0.9051, 0.805]

ax.plot(only_kernel, marker='o', markersize=30, color='black', linewidth=4, linestyle='--')
ax.plot(only_shift, marker='+', markersize=30, color='black', linewidth=4, linestyle=':')
ax.plot(shfit_and_fft, marker='x', markersize=30, color='black', linewidth=4, linestyle='-.')
ax.plot(combined, marker='*', markersize=30, color='black', linewidth=4, linestyle='-')

tick_names = ['160 \n (Baseline)', 128, 96, 64, 32]
ax.set_xticks(range(len(tick_names)))
ax.set_xticklabels(tick_names)

#enlarge tick font size
ax.tick_params(axis='both', which='major', labelsize=60)

#set ylim
ax.set_ylim(0.75, 0.95)
ax.grid(True, axis='y', linestyle='--', linewidth=0.5, color='black', alpha=0.3)

# add y axis name
ax.set_ylabel('Note onset F1 score', fontsize=60)
ax.set_xlabel('Latency (ms)', fontsize=60)

# add legend
ax.legend(['Kernel size', 'Label shift', 'Window + shift', 'Combined best'], loc='lower left', fontsize=50)
```
